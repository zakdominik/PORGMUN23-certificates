## Data Importing
Loads excel sheet into a pandas dataframe. Removes all NaN values. Make all names start with a capital letter, other characters lowercase.

```python
def import_data(file, sheet):
    raw_data = pd.read_excel(file, sheet_name=sheet)
    raw_data2 = raw_data[["Name", "School", "Final position"]]
    pd.set_option('display.max_rows', None)
    data=raw_data2.dropna()
    data["Name"] = data["Name"].str.lower()
    data["Name"] = data["Name"].str.title()
    data.reset_index(drop=True, inplace=True)
    display(data)
    return data

data = import_data("Registration_2023.xlsx", "Delegates")
```

## Data Handling
Saves columns of the Dataframe into lists. Modifies the lists so that there are no special characters and it is in the same format. Additionally, committee abbreviations are changed into full names.

```python
def data_handling(df):
    names = data["Name"].tolist()
    schools = data["School"].tolist()
    final_position = data["Final position"].tolist()
    committees = []
    countries = []

    for x in final_position:
        position = x
        parsed = []
        if "-" in position:
            position = position.replace("-", ",")
        if "," in position:
            parsed = position.split(",")
        else:
            if "un women" in position.lower():
                parsed_women = position.split(" ", 2)
                committee = parsed_women[0] + " " + parsed_women[1]
                parsed.append(committee)
                parsed.append(parsed_women[2])
            elif "icj" in position.lower():
                parsed.append("ICJ")
                parsed.append("Judge")
            elif "boxer protocol" in position.lower():
                parsed_boxer = position.split(" ", 2)
                committee = parsed_boxer[0] + " " + parsed_boxer[1]
                parsed.append(committee)
                parsed.append(parsed_boxer[2])
            else:
                parsed = position.split(" ")
        committees.append(parsed[0])
        countries.append(parsed[1])

    #abr to full names
    committees_abr = ["BOXER PROTOCOL", "ECOSOC", "HRC", "SC", "UNODC", "DISEC", "AL", "AU", "ICJ", "SPECPOL", "UN WOMEN", "HSC"]
    committees_full = ["Boxer Protocol", "Economic and Social Council", "Human Rights Council", "Security Council", "UN Office on Drugs and Crime", "Disarmament and International Security Council", "Arab League", "African Union", "International Court of Justice", "Special Political and Decolonization Committee", "UN Women Committee", "Historical Security Council"]

    for x in range(0, len(committees)):
        for y in range(0, len(committees_abr)):
            if committees[x].upper() == committees_abr[y]:
                committees[x] = committees_full[y]
    return names, schools, committees, countries

names, schools, committees, countries = data_handling(data)

for x in range(0, len(countries)):
    countries[x] = countries[x].strip()
    if "usa" in countries[x].lower():
        countries[x] = "United States of America"
    if "uk" in countries[x].lower():
        countries[x] = "United Kingdom"
    if "ussr" in countries[x].lower():
        countries[x] = "Soviet Union"
    if "turkey" in countries[x].lower():
        countries[x] = "Türkiye"
```

```python
#loads excel, creates a list of all columns
raw_countries_df = pd.read_excel("countries_list.xlsx")
list_columns = list(raw_countries_df)
for x in list_columns:
    if "Unnamed" in x:
        list_columns.remove(x)

#puts all columns into one list
df_list = []
for x in list_columns:
    df_list.append(raw_countries_df[x].tolist())

#boxer protocol with nan values
boxer_protocol_nan = []
boxer_protocol_nan.append(df_list[10])
boxer_protocol_nan.append(df_list[11])

#historical council with nan values
historical_council_nan = df_list[0]

#normal councils with nan values
others_nan = []

for x in range(1, 11):
    for y in df_list[x]:
        others_nan.append(y)
for x in df_list[12]:
    others_nan.append(x)
#check if really 1:10 and 12??

historical_council = []
others = []

for x in others_nan:
    if str(x) != "nan":
        others.append(x)

for x in historical_council_nan:
    if str(x) != "nan":
        historical_council.append(x)

#do the same for boxer, divides boxer protocol into name and country lists
boxer_name = []
boxer_country = []

for x in boxer_protocol_nan[0]:
    if str(x) != "nan":
        boxer_name.append(x)

for x in boxer_protocol_nan[1]:
    if str(x) != "nan":
        boxer_country.append(x)
        
print(boxer_name)
print(boxer_country)
boxer_name_ordered = []

#.lower!
for x in range(0, len(countries)):
    if committees[x] == "Historical Security Council":
        for y in historical_council:
            if countries[x].lower() in y.lower():
                countries[x] = y
    elif committees[x] == "Boxer Protocol":
        for y in boxer_country:
            if countries[x].lower() in y.lower():
                countries[x] = y
                # append name to boxer_name_ordered so that it can be later used
    else:
        for y in others:
            if countries[x].lower() in y.lower():
                countries[x] = y
    
print()
#creates a new list where the boxer protocol names are ordered as they go in the table
boxer_names_ordered = []

for x in range(0, len(countries)):
    if committees[x].lower() == "boxer protocol":
        country = countries[x]
        for y in boxer_country:
            if country.lower() in y.lower():
                country = y
        index = boxer_country.index(country)
        boxer_names_ordered.append(boxer_name[index])

print(boxer_names_ordered)
```

## Certificates
Creates a custom certificate for every delegate. Includes their name, country, committee. Saves them to a folder called certificates <br>TO DO: Official countries names

```python
def create_certificates(names, committees, countries, number_of_certificates):
    roboto_normal = ImageFont.truetype("Roboto-Regular.ttf", 72)
    roboto_title = ImageFont.truetype("Roboto-Bold.ttf", 125)
    roboto_1 = ImageFont.truetype("Roboto-Bold.ttf", 72)
    roboto_2 = ImageFont.truetype("Roboto-Bold.ttf", 90)
    roboto_uk = ImageFont.truetype("Roboto-Bold.ttf", 65)
    roboto_uk_normal = ImageFont.truetype("Roboto-Regular.ttf", 65)
    roboto_bottom_1 = ImageFont.truetype("Roboto-Bold.ttf", 55)
    roboto_bottom_2 = ImageFont.truetype("Roboto-Regular.ttf", 55)

    i = 0
    boxer = 0
    
    for x in range(0, number_of_certificates):
        image = Image.open("certificate_final.jpg")
        width, height = image.size
        certificate = ImageDraw.Draw(image)
        roboto_normal = ImageFont.truetype("Roboto-Regular.ttf", 72)
        #certificate.line(((0, 686.5), (1576, 686.5)), "gray")
        #certificate.line(((1250, 0), (1250, 2500)), "gray")
        certificate.text((1250, 640), names[x], (0, 0, 0), font=roboto_2, anchor="mb")
        certificate.text((1250, 300), "CERTIFICATE OF PARTICIPATION", (0, 0, 0), font=roboto_title, anchor="mb")
        certificate.text((1250, 470), "this is to certify that", (0, 0, 0), font=roboto_normal, anchor="mb")
        certificate.text((1250, 830), "has successfully participated in the", (0, 0, 0), font=roboto_normal, anchor="mb")
        certificate.text((1250, 980), "PORG MODEL UNITED NATIONS CONFERENCE 2023", (0, 0, 0), font=roboto_1, anchor="mb")

        image_2 = Image.new('RGB', (2500, 1776), (255, 255, 255))
        draw = ImageDraw.Draw(image_2)

        text_1 = ""
        text_2 = countries[x]

        if countries[x].upper() == "JUDGE":
            text_1 = "as a"
            text_size_1 = draw.textsize(text_1, font=roboto_normal)
            text_size_2 = draw.textsize(text_2, font=roboto_1)
            size = text_size_1[0] + text_size_2[0]
            one_tail = math.floor((width-size)/2)
            certificate.text((one_tail, 1080), "as a", (0, 0, 0), font=roboto_normal)
            certificate.text((one_tail + text_size_1[0] + 10, 1080), countries[x], (0, 0, 0), font=roboto_1)
        elif committees[x].lower() == "boxer protocol":
            if countries[x].lower() == "united kingdom of great britain and ireland":
                text_1 = "as"
                name = boxer_names_ordered[boxer]
                boxer += 1
                text_3 = "of the"
                text_size_1 = draw.textsize(text_1, font=roboto_uk_normal)
                name_size = draw.textsize(name, font=roboto_uk)
                text_size_3 = draw.textsize(text_3, font=roboto_uk_normal)
                country_size = draw.textsize(text_2, font=roboto_uk)
                size = text_size_1[0] + name_size[0] + text_size_3[0] + country_size[0]
                one_tail = math.floor((width-size)/2)
                certificate.text((one_tail, 1080), "as", (0, 0, 0), font=roboto_uk_normal)
                certificate.text((one_tail + text_size_1[0] + 20, 1080), name, (0, 0, 0), font=roboto_uk)
                certificate.text((one_tail + text_size_1[0] + 20 + name_size[0] + 15, 1080), "of the", (0, 0, 0), font=roboto_uk_normal)
                certificate.text((one_tail + text_size_1[0] + 20 + name_size[0] + 15 + text_size_3[0] + 10, 1080), text_2, (0, 0, 0), font=roboto_uk)
            else:
                text_1 = "as"
                name = boxer_names_ordered[boxer]
                boxer += 1
                text_3 = "of the"
                text_size_1 = draw.textsize(text_1, font=roboto_normal)
                name_size = draw.textsize(name, font=roboto_1)
                text_size_3 = draw.textsize(text_3, font=roboto_normal)
                country_size = draw.textsize(text_2, font=roboto_1)
                size = text_size_1[0] + name_size[0] + text_size_3[0] + country_size[0]
                one_tail = math.floor((width-size)/2)
                certificate.text((one_tail, 1080), "as", (0, 0, 0), font=roboto_normal)
                certificate.text((one_tail + text_size_1[0] + 20, 1080), name, (0, 0, 0), font=roboto_1)
                certificate.text((one_tail + text_size_1[0] + 20 + name_size[0] + 15, 1080), "of the", (0, 0, 0), font=roboto_normal)
                certificate.text((one_tail + text_size_1[0] + 20 + name_size[0] + 15 + text_size_3[0] + 10, 1080), text_2, (0, 0, 0), font=roboto_1)
        else:    
            if countries[x].lower() == "united kingdom of great britain and northern ireland":
                text_1 = "as the delegate of"
                text_size_1 = draw.textsize(text_1, font=roboto_uk_normal)
                text_size_2 = draw.textsize(text_2, font=roboto_uk)
                size = text_size_1[0] + text_size_2[0]
                one_tail = math.floor((width-size)/2)
                certificate.text((one_tail, 1080), "as the delegate of", (0, 0, 0), font=roboto_uk_normal)
                certificate.text((one_tail + text_size_1[0] + 10, 1080), countries[x], (0, 0, 0), font=roboto_uk)
            else:
                text_1 = "as the delegate of"
                text_size_1 = draw.textsize(text_1, font=roboto_normal)
                text_size_2 = draw.textsize(text_2, font=roboto_1)
                size = text_size_1[0] + text_size_2[0]
                one_tail = math.floor((width-size)/2)
                certificate.text((one_tail, 1080), "as the delegate of", (0, 0, 0), font=roboto_normal)
                certificate.text((one_tail + text_size_1[0] + 10, 1080), countries[x], (0, 0, 0), font=roboto_1)


        text_3 = "in the"
        text_size_3 = draw.textsize(text_3, font=roboto_normal)
        text_4 = committees[x]
        text_size_4 = draw.textsize(text_4, font=roboto_1)
        size_2 = text_size_3[0] + text_size_4[0]
        one_tail_2 = math.floor((width-size_2)/2)

        certificate.text((one_tail_2, 1230), text_3, (0, 0, 0), font=roboto_normal)
        certificate.text((one_tail_2 + text_size_3[0] + 20, 1230), committees[x], (0, 0, 0), font=roboto_1)

        certificate.text((210, 1580), "Secretary General", (0, 0, 0), font=roboto_bottom_1)
        certificate.text((210, 1510), "Andrea Kutnarová", (0, 0, 0), font=roboto_bottom_2)

        certificate.text((1940, 1580), "Supervisor", (0, 0, 0), font=roboto_bottom_1)
        certificate.text((1920, 1510), "Zach Barnes", (0, 0, 0), font=roboto_bottom_2)


        image_name = "certificate" + str(i) + ".jpg"
        image.save("certificates/" + image_name)
        i = i+1

    
create_certificates(names, committees, countries, len(names)) #len(names) !! check if correct number

```
